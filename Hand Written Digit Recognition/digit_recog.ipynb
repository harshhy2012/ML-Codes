```python
import tensorflow as tf 
import matplotlib.pyplot as plt
import numpy as np 
```

```python
#importing dataset
mnist = tf.keras.datasets.mnist
```

```python
#spliting test and train data
(x_train,y_train), (x_test,y_test) = mnist.load_data()
x_train =  tf.keras.utils.normalize(x_train, axis=1)
x_test = tf.keras.utils.normalize(x_test, axis=1)
#visualizing the data
plt.imshow(x_train[0],cmap=plt.cm.binary)
plt.show()
print(x_train[0])
```

```python
#visualizing the data
plt.imshow(x_train[0],cmap=plt.cm.binary)
plt.show()
print(x_train[0])
```

```python
#making and training neural network
#choosing type of NN model
model = tf.keras.models.Sequential()
#adding layers to model (1 flatten, 3 Dense)
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(128, activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(128, activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(10, activation=tf.nn.softmax))
#choosing optimization algorith, loss function and metrics to be printed after every epoch
model.compile(optimizer='adam' , loss='sparse_categorical_crossentropy',metrics=['accuracy'])
#training
model.fit(x_train,y_train,epochs=3)
```

```python
#if we directly print prediction we get a multidimentional numpy array as output
predictions = model.predict(x_test)
print(predictions)
```

```python
#if we print using np.argmax function we get the prediction as a integer number
print(type(predictions))
print(np.argmax(predictions[3]))
```
