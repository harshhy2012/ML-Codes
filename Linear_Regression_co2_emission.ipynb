```python
#Reading the data from the downloaded csv
df = pd.read_csv("FuelConsumption.csv")
#taking a look at the datafram
df.head()
```

```python
# summarize the data
df.describe()
```

```python
#Selecting relevent data for linear regression
cdf = df[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB','CO2EMISSIONS']]
cdf.head(9)
```

```python
viz = cdf[['CYLINDERS','ENGINESIZE','CO2EMISSIONS','FUELCONSUMPTION_COMB']]
viz.hist()
plt.show()
```

```python
plt.scatter(cdf.FUELCONSUMPTION_COMB, cdf.CO2EMISSIONS, COLOR = 'red')
plt.xlabel('FUELCONSUMPTION_COMB')
plt.ylabel('Emission')
plt.show()
```

```python
plt.scatter(cdf.ENGINESIZE, cdf.CO2EMISSIONS, COLOR='red')
plt.xlabel('Engine Size')   
plt.ylabel('Emission')
```

```python
plt.scatter(cdf.CYLINDERS, cdf.CO2EMISSIONS, COLOR='red')
plt.xlabel('Cylinders')   
plt.ylabel('Emission')
```

```python
#Splitting into test and train dataframes
msk = np.random.rand(len(df)) < 0.8
train = cdf[msk]
test = cdf[~msk]
```

```python
plt.scatter(train.ENGINESIZE, train.CO2EMISSIONS, COLOR = 'Blue')
plt.xlabel("Engine Size")
plt.ylabel("Emission")
plt.show()
```

```python
#Modeling
#import linear regression model from sk-learn
from sklearn import linear_model
regr = linear_model.LinearRegression()
#make np arrays for dependent and independent variables in the training dataframe
train_x = np.asanyarray(train[['ENGINESIZE']])
train_y = np.asanyarray(train[['CO2EMISSIONS']])
#fit the parameters for your model
regr.fit(train_x,train_y)
print ('Coefficients: ', regr.coef_)
print ('Intercept: ',regr.intercept_)
```

```python
#plotting regression line of our model
plt.scatter(train.ENGINESIZE, train.CO2EMISSIONS,  color='blue')
plt.plot(train_x, regr.coef_[0][0]*train_x + regr.intercept_[0], '-r')
plt.xlabel("Engine size")
plt.ylabel("Emission")
```

```python
#evaluation of the model
#import r2 from sk-learn
from sklearn.metrics import r2_score
#make np arrays for dependent and independent variables in the test dataframe
test_x = np.asanyarray(test[['ENGINESIZE']])
test_y = np.asanyarray(test[['CO2EMISSIONS']])
#predict for test_x
test_y_hat = regr.predict(test_x)

print("Mean absolute error: %.2f" % np.mean(np.absolute(test_y_hat - test_y)))
print("Residual sum of squares (MSE): %.2f" % np.mean((test_y_hat - test_y) ** 2))
print("R2-score: %.2f" % r2_score(test_y_hat , test_y) )
```
